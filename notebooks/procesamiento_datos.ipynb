```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
antioquia = "/content/CNPV2018_1VIV_A2_05.CSV"
data_ant = pd.read_csv(antioquia)
bogota = "/content/CNPV2018_1VIV_A2_11.CSV"
data_bog = pd.read_csv(bogota)
```

```python
# Verificar si hay datos faltantes
print("datos faltantes antioquia")
print(data_ant.isnull().sum())
print("datos faltantes bogota")
print(data_bog.isnull().sum())



print("total de datos antioquia: ",len(data_ant))

print("total de datos bogota: ",len(data_bog))
```

Podemos ver que hay varias columnas con mas del 50% de datos en NULL, estas columnas no nos servir√°n de mucho para visualizar los datos, entonces procedemos a eliminarlas


```python

data_ant.drop("UVA1_TIPOTER", axis=1, inplace=True)
data_ant.drop("UVA2_CODTER", axis=1, inplace=True)
data_ant.drop("UVA1_COD_AREAPROT", axis=1, inplace=True)
data_ant.drop("L_TIPO_INST", axis=1, inplace=True)
data_ant.drop("L_EXISTEHOG", axis=1, inplace=True)
data_ant.drop("L_TOT_PERL", axis=1, inplace=True)
data_bog.drop("UVA_ESTATER", axis=1, inplace=True)
data_bog.drop("UVA1_TIPOTER", axis=1, inplace=True)
data_bog.drop("UVA2_CODTER", axis=1, inplace=True)
data_bog.drop("UVA1_COD_AREAPROT", axis=1, inplace=True)
data_bog.drop("L_TIPO_INST", axis=1, inplace=True)
data_bog.drop("L_EXISTEHOG", axis=1, inplace=True)
data_bog.drop("L_TOT_PERL", axis=1, inplace=True)
```

```python
import statistics

#----------reemplazo por la moda en las variables con null de ANTIOQUIA
v_tipo_viv_moda = statistics.mode(data_ant['V_TIPO_VIV'])
data_ant['V_TIPO_VIV'] = data_ant['V_TIPO_VIV'].fillna(v_tipo_viv_moda)

v_con_ocup_moda = statistics.mode(data_ant['V_CON_OCUP'])
data_ant['V_CON_OCUP'] = data_ant['V_CON_OCUP'].fillna(v_con_ocup_moda)

v_tot_hog_moda = statistics.mode(data_ant['V_TOT_HOG'])
data_ant['V_TOT_HOG'] = data_ant['V_TOT_HOG'].fillna(v_tot_hog_moda)

v_tot_hog_moda = statistics.mode(data_ant['V_TOT_HOG'])
data_ant['V_TOT_HOG'] = data_ant['V_TOT_HOG'].fillna(v_tot_hog_moda)

v_mat_pared_moda = statistics.mode(data_ant['V_MAT_PARED'])
data_ant['V_MAT_PARED'] = data_ant['V_MAT_PARED'].fillna(v_mat_pared_moda)

v_mat_piso_moda = statistics.mode(data_ant['V_MAT_PISO'])
data_ant['V_MAT_PISO'] = data_ant['V_MAT_PISO'].fillna(v_mat_piso_moda)

va_ee_moda = statistics.mode(data_ant['VA_EE'])
data_ant['VA_EE'] = data_ant['VA_EE'].fillna(va_ee_moda)

va1_estrato_moda = statistics.mode(data_ant['VA1_ESTRATO'])
data_ant['VA1_ESTRATO'] = data_ant['VA1_ESTRATO'].fillna(va1_estrato_moda)

vb_acu_moda = statistics.mode(data_ant['VB_ACU'])
data_ant['VB_ACU'] = data_ant['VB_ACU'].fillna(vb_acu_moda)

vc_alc_moda = statistics.mode(data_ant['VC_ALC'])
data_ant['VC_ALC'] = data_ant['VC_ALC'].fillna(vc_alc_moda)

vd_gas_moda = statistics.mode(data_ant['VD_GAS'])
data_ant['VD_GAS'] = data_ant['VD_GAS'].fillna(vd_gas_moda)

ve_recbas_moda = statistics.mode(data_ant['VE_RECBAS'])
data_ant['VE_RECBAS'] = data_ant['VE_RECBAS'].fillna(ve_recbas_moda)

ve1_qsem_moda = statistics.mode(data_ant['VE1_QSEM'])
data_ant['VE1_QSEM'] = data_ant['VE1_QSEM'].fillna(ve1_qsem_moda)

vf_internet_moda = statistics.mode(data_ant['VF_INTERNET'])
data_ant['VF_INTERNET'] = data_ant['VF_INTERNET'].fillna(vf_internet_moda)

v_tipo_sersa_moda = statistics.mode(data_ant['V_TIPO_SERSA'])
data_ant['V_TIPO_SERSA'] = data_ant['V_TIPO_SERSA'].fillna(v_tipo_sersa_moda)


#----------reemplazo por la moda en las variables con null de BOGOTA

v_tipo_viv_moda = statistics.mode(data_bog['V_TIPO_VIV'])
data_bog['V_TIPO_VIV'] = data_bog['V_TIPO_VIV'].fillna(v_tipo_viv_moda)

v_con_ocup_moda = statistics.mode(data_bog['V_CON_OCUP'])
data_bog['V_CON_OCUP'] = data_bog['V_CON_OCUP'].fillna(v_con_ocup_moda)

v_tot_hog_moda = statistics.mode(data_bog['V_TOT_HOG'])
data_bog['V_TOT_HOG'] = data_bog['V_TOT_HOG'].fillna(v_tot_hog_moda)

v_tot_hog_moda = statistics.mode(data_bog['V_TOT_HOG'])
data_bog['V_TOT_HOG'] = data_bog['V_TOT_HOG'].fillna(v_tot_hog_moda)

v_mat_pared_moda = statistics.mode(data_bog['V_MAT_PARED'])
data_bog['V_MAT_PARED'] = data_bog['V_MAT_PARED'].fillna(v_mat_pared_moda)

v_mat_piso_moda = statistics.mode(data_bog['V_MAT_PISO'])
data_bog['V_MAT_PISO'] = data_bog['V_MAT_PISO'].fillna(v_mat_piso_moda)

va_ee_moda = statistics.mode(data_bog['VA_EE'])
data_bog['VA_EE'] = data_bog['VA_EE'].fillna(va_ee_moda)

va1_estrato_moda = statistics.mode(data_bog['VA1_ESTRATO'])
data_bog['VA1_ESTRATO'] = data_bog['VA1_ESTRATO'].fillna(va1_estrato_moda)

vb_acu_moda = statistics.mode(data_bog['VB_ACU'])
data_bog['VB_ACU'] = data_bog['VB_ACU'].fillna(vb_acu_moda)

vc_alc_moda = statistics.mode(data_bog['VC_ALC'])
data_bog['VC_ALC'] = data_bog['VC_ALC'].fillna(vc_alc_moda)

vd_gas_moda = statistics.mode(data_bog['VD_GAS'])
data_bog['VD_GAS'] = data_bog['VD_GAS'].fillna(vd_gas_moda)

ve_recbas_moda = statistics.mode(data_bog['VE_RECBAS'])
data_bog['VE_RECBAS'] = data_bog['VE_RECBAS'].fillna(ve_recbas_moda)

ve1_qsem_moda = statistics.mode(data_bog['VE1_QSEM'])
data_bog['VE1_QSEM'] = data_bog['VE1_QSEM'].fillna(ve1_qsem_moda)

vf_internet_moda = statistics.mode(data_bog['VF_INTERNET'])
data_bog['VF_INTERNET'] = data_bog['VF_INTERNET'].fillna(vf_internet_moda)

v_tipo_sersa_moda = statistics.mode(data_bog['V_TIPO_SERSA'])
data_bog['V_TIPO_SERSA'] = data_bog['V_TIPO_SERSA'].fillna(v_tipo_sersa_moda)

```
